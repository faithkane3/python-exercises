1. Use matplotlib to plot the following equation:



y = $x^2$ - x + 2
 
Add an anotation for the point 0, 0, the origin.

```python
import matplotlib.pyplot as plt
import math
import random
import numpy as np
%matplotlib inline
```

```python
x = range(-10, 10, 1)
y = [i ** 2 - i + 2 for i in x]

plt.plot(x,y, c = "blueviolet")
plt.title("y = $x^2$ - x + 2")
plt.annotate('(0,0)', xy=(0, 0), xytext=(-3, 5),
             arrowprops={'facecolor': 'firebrick'})
plt.show()
```

3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.figure(figsize=(20, 10))

n_rows = 2
n_cols = 2

# the first subplot
plt.subplot(n_rows, n_cols, 1)

x = range(0,100, 1)
y = [math.sqrt(i) for i in x]

plt.plot(x,y)
plt.title("y = $\sqrt{x}$")

# the second subplot
plt.subplot(n_rows, n_cols, 2)

x = np.arange(-10,10,1)
y = [i ** 3 for i in x]

plt.plot(x,y, c="orange")
plt.title("y = $x^3$")

# the third subplot
plt.subplot(n_rows, n_cols, 3)

x = np.arange(-80,80, .1)
y = [math.tan(i) for i in x]

plt.plot(x,y, c="purple")
plt.title("y = tan(x)")

# the fourth subplot
plt.subplot(n_rows, n_cols, 4)

x = np.arange(0, 30)
y = [2 ** i for i in x]

plt.plot(x,y, c="red")
plt.title("y = $2^x$")

plt.suptitle('Four in One')
plt.show()
```

Comine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend.

```python
x = range(0,10)
y1 = [math.sqrt(i) for i in x]
y2 = [i ** 3 for i in x]
y3 = [math.tan(i) for i in x]
y4 = [2 ** i for i in x]

plt.plot(x,y1, c="blue", label='y = $\sqrt{x}$')
plt.plot(x,y2, c="yellow", label='y = $x^3$')
plt.plot(x,y3, c="purple", label='y = tan(x)')
plt.plot(x,y4, c="red", label='y = $2^x$')
plt.annotate('Check out this mess', xy=(4,150), xytext=(0, 200),
             arrowprops={'facecolor': 'firebrick'})
plt.legend(loc='upper left')
plt.xlabel('$x$')
plt.title("The Monster")
plt.show()
```
