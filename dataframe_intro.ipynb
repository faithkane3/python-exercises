1. Copy the code from the lesson to create a dataframe full of student grades.



```python
np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})

type(df)
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

a. Create a column named passing_english that indicates whether each student has a passing grade in reading.


```python
df["passing_english"] = df.english >= 70
df
```

b. Sort the english grades by the passing_english column. How are duplicates handled?


```python
df.sort_values(by="english")
```

c. Sort the english grades first by passing_english and then by student name. All the students that are failing english should be first, and within the students that are failing english they should be ordered alphabetically. The same should be true for the students passing english. (Hint: you can pass a list to the .sort_values method)


```python
df.sort_values(by=["english", "name"])
```

d. Sort the english grades first by passing_english, and then by the actual english grade, similar to how we did in the last step.


```python
df.sort_values(by=["passing_english", "english"])
```

e. Calculate each students overall grade and add it as a column on the dataframe. The overall grade is the average of the math, english, and reading grades.


```python
df["average"] = round((df.math + df.english + df.reading) / 3, 2)
df
```

-How many animals are hoppers that are above the median speed? What percentage is this?

```python
# median speed of mammals
mam_med = Mammals.speed.median()
mam_med
```

```python
# median speed of mammals
above_median = Mammals[Mammals.speed > Mammals.speed.median()].head(5)
above_median
```

```python
# number of mammals with speeds above median speed
num_above_median = len(Mammals[Mammals.speed > Mammals.speed.median()].sort_values(by="speed", ascending=False))
num_above_median
```

```python
num_above_median / len(Mammals) * 100
```

### same as above but filtering for hoppers

```python
# median speed of mammals that are hoppers
hoppers_med_speed = Mammals.speed[Mammals["hoppers"] == True].median()
hoppers_med_speed
```

```python
# subset of mammals that are hoppers
hoppers = Mammals[Mammals["hoppers"] == True]
hoppers
```

```python
# Mammals who are hoppers with speed above median hopper speed 
hoppers_above_hop_med = hoppers[Mammals.speed[Mammals["hoppers"] == True] > mam_med]
hoppers_above_hop_med
```

```python
percent_hoppers_above_med = len(hoppers_above_hop_med) / mam_med * 100
percent_hoppers_above_med
```

```python
num_hoppers = sum((Mammals.hoppers == True) & (Mammals.speed > Mammals.speed.median()))
num_hoppers
```

```python
print(round(num_hoppers/len(Mammals) * 100,2))
```
